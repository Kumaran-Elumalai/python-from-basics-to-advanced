### Difference Between append and extend

This is a very common interview question.

**append()**
- Adds the entire object as **one single element**.
- The list grows by **1 element**, no matter what you append.

Example:  
`my_list.append([4, 5])`  
Result: `[1, 2, 3, [4, 5]]`

**extend()**
- Adds each element of the object **individually**.
- The list grows by the **number of elements** in the iterable.

Example:  
`my_list.extend([4, 5])`  
Result: `[1, 2, 3, 4, 5]`

### In simple words:
- **append → adds the box**  
- **extend → opens the box and adds the items inside**

## Lists in Memory

How lists store references, not actual values\
A list `doesn’t store the actual data`. It `stores references (pointers)` to the data.\
So when you do

````python
a = [10, 20, 30]
````

Python stores three numbers somewhere in memory, and the list simply points to them.\
Understanding this explains many surprising list behaviors.

**Ids and pointers illustrated**

````python
x = 10
y = 10
id(x), id(y)
````

Both may show the same id because Python reuses small integers.

For lists:

````python
a = [10]
id(a[0])    # shows the id of the actual object
````

##### The list just holds these ids behind the scenes.

Why list multiplication [*3] can cause bugs\
This is a common trap.

````python
a = [[0] * 3] * 3
````

````python
print(a)
````

Looks like three separate rows, but all rows point to the same inner list.\
Changing one row changes all.

````python
a[0][0] = 99
print(a)
````

The bug happens because each row shares the same reference.

#### Shallow vs Deep Copy
A shallow copy creates a new outer list but still points to the same inner elements.

**copy()**

````python
a = [1, 2, 3]
b = a.copy()
print("a: ",a)
print("b: ",b)
````

Good for simple lists with no nested structures.

**Slicing**

````python
b = a[:]
print("a: ",a)
print("b: ",b)
````

Works like `copy().` Still shallow.

**Deep Copy**
To fully copy everything, including nested lists:

````python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

a[0][0] = 99
print("a: ",a)
print("b: ",b)
````

Now changes to a don’t affect b because every level is independently copied.

This is the safest option when working with nested lists.

#### Flattening deeply nested lists
A common data-cleaning task.

````python
def flatten(nested):
    flat = []
    for item in nested:
        if isinstance(item, list):
            flat.extend(flatten(item))
        else:
            flat.append(item)
    return flat

data = [1, [2, [3, 4], 5], 6]

flatten(data)
````

## Introduction to Tuples
 
A tuple is an ordered collection of values similar to a list, with one major difference:  
Tuples are **immutable**, which means you cannot modify them after creation.


### Why Do Tuples Exist?

Tuples are useful because they offer several advantages:

- Immutability provides safety since values cannot be changed accidentally  
- Faster performance compared to lists  
- Ideal for storing fixed sets of data such as coordinates, dates, or configuration values  
- Tuples are hashable, so they can be used as dictionary keys  
  (lists cannot be used as dictionary keys)

### Tuple vs List

| Feature      | List                          | Tuple                              |
|--------------|--------------------------------|-------------------------------------|
| Mutability   | Mutable (can be modified)      | Immutable (cannot be modified)      |
| Syntax       | `[1, 2, 3]`                    | `(1, 2, 3)`                         |
| Speed        | Slower                         | Faster                              |
| Use-case     | When data needs to change      | When data should remain fixed       |
| Hashable     | Not hashable                   | Hashable (can be used as dict keys) |

# Memory and Performance

Understanding how tuples behave under the hood shows real depth.  
This is the kind of knowledge that **impresses recruiters**.


### 1. Why Tuples Are Faster Than Lists

Tuples are **immutable**, so Python doesn’t need to:

- track **changes**
- allocate extra space for **future growth**
- maintain modification **methods**

Because of this, Python can **store** and **access** tuple elements more efficiently.

**In short:**  
**Less flexibility = more speed**

### 2. Tuple’s Memory Efficiency

Lists are designed to grow, so Python allocates **extra unused space** inside them.

Tuples never change, so Python stores **only the exact number of items** with no additional overhead.

**Example thought:**

- A list with 5 items may internally reserve space for **8 or 10**.
- A tuple with 5 items stores **exactly 5**.

This makes tuples:

- **Smaller in memory footprint**
- **More cache-friendly**

### 3. Interning and Optimization

Python sometimes internally **reuses immutable objects** to save memory — a technique called **interning**.

Tuples benefit from this because:

- They are **immutable**
- They often store **constant values**
- They frequently appear in **function signatures**, **loops**, and **class definitions**

Because of this, the Python interpreter can **optimize and reuse** tuple objects instead of creating new ones.

**Key Insight:**  
Tuples that contain **only immutable items** (such as integers or strings) can sometimes be **interned or reused internally**, reducing memory usage and improving performance.

### 4. Where Python Prefers Tuples Internally

Python uses **tuples behind the scenes** in many core operations because they are **fixed**, **lightweight**, and **fast**.

Here are common places where Python internally relies on tuples:

- **Function arguments** are packed into a tuple.
- **Multiple assignment** (e.g., `x, y, z = ...`) uses **tuple packing and unpacking**.
- The **range object** uses tuples for some internal calculations.
- **Constant tuples** inside your code may be created **once** and then **reused**.
- The **Abstract Syntax Tree (AST)** stores many internal nodes as tuples.

**Key idea:**  
Whenever Python needs a **fixed**, **read-only**, and **efficient** group of values, it chooses a **tuple** automatically.
